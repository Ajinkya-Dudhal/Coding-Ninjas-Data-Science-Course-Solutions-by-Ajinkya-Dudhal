```python
import matplotlib.pyplot as plt
import numpy as np

employees = np.array([61,71,79,91,93,89,90,94,99,128,118,114,124,131])
year = np.arange(2005,2019)
revenue = np.array([39.79,44.28,51.12,60.42,58.44,62.48,69.94,73.72,77.85,86.83,93.58,85.32,89.95,110.36])

color = np.arange(len(year))

plt.scatter(year,revenue,s=employees,c=color)
plt.title('Microsoft Bubble Graph')
plt.xlabel('year')
plt.ylabel('revenue')
plt.grid()
plt.show()

peak1 = np.where(year==2015)
peak2 = np.where(year==2008)
print(2008,revenue[peak2],employees[peak2])
print(2015,revenue[peak1],employees[peak1])
```

```python
import matplotlib.pyplot as plt
import numpy as np

def func(laptop,sold):
    li = []
    arr = np.array((sold*100)/np.sum(sold))
    for i in range(len(sold)):
        print(laptop[i],round(arr[i],1))
        li.append(arr[i])
    s = str(li)
    return s
    
laptop = np.array(['HP','Dell','Asus','Lenovo','Apple','Acer'])
sold = np.array([20000,43000,15000,17000,22000,13000])

plt.pie(sold,labels=laptop,autopct=func(laptop,sold))
plt.show()

# print(laptop[0],15.4)
# print(laptop[1],33.1)
# print(laptop[2],11.5)
# print(laptop[3],13.1)
# print(laptop[4],16.9)
# print(laptop[5],10.0)
```

```python
df = pd.read_csv('amazon_jobs_dataset.csv')
df = df.iloc[:,1:]
for i in range(df.shape[0]):
    df.iloc[i,2] = df.iloc[i,2].split(" ")[0]
month = df['Posting_date'].value_counts()
month = month.sort_index()
print(month)

name = []
count = []
for i in range(len(month)):
    x = month.index[i]
    name.append(str(x))
    count.append(month[x])

plt.bar(name,count)
plt.xticks(rotation=50)
plt.show()
```

```python
import csv

with open('amazon_jobs_dataset.csv',encoding='utf8') as file_obj:
    file_data = csv.DictReader(file_obj)
    dic = {} 
    for d in file_data:
        month = d['Posting_date'].split(" ")[0]
        if month in dic:
            dic[month] = dic[month] + 1
        else:
            dic[month] = 1
    print(dic)
    month = dic.keys()
    count = dic.values()
    
    plt.bar(month,count)
    plt.xticks(rotation=50)
    plt.title('Month vs Job Openings')
    plt.xlabel('Month')
    plt.ylabel('No. of Job Openings')
    plt.show()
```

```python
df = pd.read_csv('amazon_jobs_dataset.csv')
df = df.iloc[:,1:]
df['Country'] = np.nan

for i in range(df.shape[0]):
    df.iloc[i,6] = df.iloc[i,1].split(",")[0]
    df.iloc[i,1] = df.iloc[i,1].split(",")[-1]
df = df[df['Country']=='IN']
city = df['location'].value_counts()

name = []
count = []
for i in range(len(city)):
    x = city.index[i]
    name.append(x)
    count.append(city[x])

print(city)
plt.pie(count,labels=name,autopct='%.2f%%')
plt.show()
```

```python
import csv

with open('amazon_jobs_dataset.csv',encoding='utf8') as file:
    file_data = csv.DictReader(file)
    dic = {}
    for d in file_data:
        qual = d['BASIC QUALIFICATIONS']
        year = int(d['Posting_date'].split(",")[-1])
        if 'Java' in qual or 'java' in qual:
            if year in dic:
                dic[year] = dic[year] + 1
            else:
                dic[year] = 1
    print(dic)
    
    keys = dic.keys()
    values = dic.values()
    plt.scatter(keys,values)
    plt.xlabel('Year')
    plt.ylabel('No. of Jobs')
    plt.show()
```

```python
df = pd.read_csv('amazon_jobs_dataset.csv')
df['Posting_date'].apply(lambda x: x.split(",")[-1].strip())
df.drop(subset=['BASIC QUALIFICATIONS'],df['BASIC QUALIFICATIONS'].apply(lambda x: ('Java' in x) | ('java' in x)),inplace=True)
df
```
