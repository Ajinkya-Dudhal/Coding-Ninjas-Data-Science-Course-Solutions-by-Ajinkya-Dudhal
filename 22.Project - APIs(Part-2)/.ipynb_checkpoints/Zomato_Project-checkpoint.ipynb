```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Part 1
The dataset is highly skewed toward the cities included in Delhi-NCR. So, we will summarise all the other cities in Rest of India while those in New Delhi, Ghaziabad, Noida, Gurgaon, Faridabad to Delhi-NCR. Doing this would make our analysis turn toward Delhi-NCR v Rest of India.

# Problem 1.1 :
Plot the bar graph of number of restaurants present in Delhi NCR vs Rest of India.

```python
df = pd.read_csv('zomato.csv',encoding='ISO-8859-1')
df = df[df['Currency']=='Indian Rupees(Rs.)']
df_Delhi = df[(df['City']=='Gurgaon') | (df['City']=='New Delhi') | (df['City']=='Ghaziabad') | (df['City']=='Noida') | (df['City']=='Faridabad')]
Delhi_count = df_Delhi['City'].count()
df_total = df['City'].count()
rest_count = df_total - Delhi_count

x = ['Delhi NCR','Rest of India']
y = [Delhi_count,rest_count]
plt.bar(x[0],y[0],width=0.5)
plt.bar(x[1],y[1],width=0.5)
plt.title('Number of restaurants in Delhi NCR vs Rest of India')
plt.xlabel('Cities')
plt.ylabel('Number of Restaurants')
plt.legend(y)
plt.show()
```

# Problem 1.2 :
Find the cuisines which are not present in restaurant of Delhi NCR but present in rest of India.Check using Zomato API whether this cuisines are actually not served in restaurants of Delhi-NCR or just it due to incomplete dataset.

```python
df = pd.read_csv('zomato.csv',encoding='ISO-8859-1')
df = df[df['Country Code']==1]
df.dropna(subset=['Cuisines'],inplace=True)

df_delhi = df[(df['City']=='Gurgaon') | (df['City']=='New Delhi') | (df['City']=='Ghaziabad') | (df['City']=='Noida') | (df['City']=='Faridabad')]
df_restIndia = df[(df['City']!='Gurgaon') & (df['City']!='New Delhi') & (df['City']!='Ghaziabad') & (df['City']!='Noida') & (df['City']!='Faridabad')]

delhi_cuisines = {}
def func(x):
    menu = x.split(",")
    for dish in menu:
        dish = dish.strip()
        if dish not in delhi_cuisines:
            delhi_cuisines[dish] = 0

df_delhi['Cuisines'].apply(lambda x: func(x))

def func1(x):
    menu = x.split(",")
    for dish in menu:
        dish = dish.strip()
        if dish not in delhi_cuisines:
            print(dish)

df_restIndia['Cuisines'].apply(lambda x: func1(x))
print()
```

# Problem 1.3 :
Find the top 10 cuisines served by maximum number of restaurants in Delhi NCR and rest of India.

```python
df = pd.read_csv('zomato.csv',encoding='ISO-8859-1')
df = df[df['Country Code']==1]
df.dropna(subset=['Cuisines'],inplace=True)
df['City'].replace(['New Delhi','Noida','Gurgaon','Faridabad','Ghaziabad'],'Delhi NCR',inplace=True)

df_delhi = df[df['City']=='Delhi NCR']
df_restIndia = df[df['City']!='Delhi NCR']

delhi_dic = {}
def func(x):
    menu = x.split(", ")
    for dish in menu:
        if dish in delhi_dic:
            delhi_dic[dish] = delhi_dic[dish] + 1
        else:
            delhi_dic[dish] = 1
df_delhi['Cuisines'].apply(lambda x: func(x))

restIndia_dic = {}
def func1(x):
    menu = x.split(", ")
    for dish in menu:
        if dish in restIndia_dic:
            restIndia_dic[dish] = restIndia_dic[dish] + 1
        else:
            restIndia_dic[dish] = 1
df_restIndia['Cuisines'].apply(lambda x: func1(x))

def top10_cuisines(keys,values,location):
    sorted_value_index = np.argsort(values)
    sorted_value_index = np.flip(sorted_value_index)[0:10]
    print(f'******Top 10 cuisines in {location}******')
    for i in sorted_value_index:
        print(keys[i],values[i])
    print()
    return sorted_value_index
    
del_keys = list(delhi_dic.keys())
del_values = list(delhi_dic.values())
delhi_index = top10_cuisines(del_keys,del_values,'Delhi NCR')

rest_keys = list(restIndia_dic.keys())
rest_values = list(restIndia_dic.values())
restIndia_index = top10_cuisines(rest_keys,rest_values,'Rest of the India')
```

# Problem 1.4 :
Write a short detailed analysis of how cuisine served is different from Delhi NCR to Rest of India. Plot the suitable graph to explain your inference.

```python
delhi_labels = []
delhi_values = []
for i in delhi_index:
    delhi_labels.append(del_keys[i])
    delhi_values.append(del_values[i])
    
restIndia_labels = []
restIndia_values = []
for i in delhi_index:
    restIndia_labels.append(rest_keys[i])
    restIndia_values.append(rest_values[i])
    
plt.pie(delhi_values,labels=delhi_labels,autopct='%.2f%%',radius=1)
plt.title('Most popular cuisines in Delhi NCR')
plt.show()

plt.pie(restIndia_values,labels=restIndia_labels,autopct='%.2f%%',radius=1)
plt.title('Most popular cuisines in rest of the India')
plt.show()
```

# Problem 2.1.1 :

```python
df = pd.read_csv('zomato.csv',encoding='ISO-8859-1')
df = df[df['Country Code']==1]
df.dropna(subset=['Aggregate rating','Votes'],inplace=True)
data = df.groupby('Aggregate rating')['Votes'].mean()
data = data.sort_values(ascending=True)

votes = data.index
rating = data
plt.figure(figsize=(15,7))
plt.plot(data,data.index,'o-')
plt.title('Votes vs Rating',fontsize=20)
plt.xlabel('Votes',size=15)
plt.ylabel('Aggregate rating',size=15)
plt.grid()
plt.show()
```

# Problem 2.1.2 :
Restaurant serving more number of cuisines

```python
df = pd.read_csv('zomato.csv',encoding='ISO-8859-1')
df = df[df['Country Code']==1]
df.dropna(subset=['Aggregate rating','Cuisines'],inplace=True)

df['Number of Cuisines'] = df['Cuisines'].apply(lambda x: len(x.split(", ")))
df_cuisines = df.groupby('Number of Cuisines')['Aggregate rating'].mean()

plt.plot(df_cuisines.index,df_cuisines,'o-')
plt.title('Number of Cuisines vs Rating',fontsize=20)
plt.xlabel('Number of Cuisines',size=15)
plt.ylabel('Rating',size=15)
plt.grid()
plt.show()
```

# Problem 2.1.4 :
Restaurant serving some specific cuisines

```python
df = pd.read_csv('zomato.csv',encoding='ISO-8859-1')
df = df[df['Country Code']==1]
df.dropna(subset=['Cuisines','Aggregate rating'],inplace=True)

dic = {}
def func(x,y):
    menu = x.split(", ")
    for dish in menu:
        if dish in dic:
            dic[dish].append(y)
        else:
            dic[dish] = [y]
            
for i in range(df.shape[0]):
    func(df.iloc[i,9],df.iloc[i,17])
for dish in dic:
    dic[dish] = round(np.mean(dic[dish]),2)

keys = list(dic.keys())
values = list(dic.values())
ascending_value_index = np.argsort(values)
descending_value_index = np.flip(ascending_value_index)
top10 = descending_value_index[0:10]
low10 = ascending_value_index[0:10]

print('****Top 10 rated cuisines served****')
cuisines = []
rating = []
for i in top10:
    print(keys[i],values[i])
    cuisines.append(keys[i])
    rating.append(values[i])
print()

plt.bar(cuisines,rating,width=0.5)
plt.xticks(rotation=50)
plt.title('Top 10 rated cuisines served',fontsize=20)
plt.xlabel('Cuisines',size=15)
plt.ylabel('Rating',size=15)
plt.show()
    
print('****Lowest 10 rated cuisines served****')
cuisines = []
rating = []
for i in low10:
    print(keys[i],values[i])
    cuisines.append(keys[i])
    rating.append(values[i])

plt.bar(cuisines,rating,width=0.5)
plt.xticks(rotation=50)
plt.title('Lowest 10 rated cuisines served',fontsize=20)
plt.xlabel('Cuisines',size=15)
plt.ylabel('Rating',size=15)
plt.show()
```

# Problem 2.2 :
Find the weighted restaurant rating of each locality and find out the top 10 localities with more weighted restaurant rating?

Weighted Restaurant Rating= (number of votes * rating) /  (number of votes) .

```python
df = pd.read_csv('zomato.csv',encoding='ISO-8859-1')
df = df[df['Country Code']==1]
df.dropna(subset=['Locality','Aggregate rating','Votes'],inplace=True)

x = df.groupby('Locality').apply(lambda x: (x['Votes']*x['Aggregate rating']).sum())
y = df.groupby('Locality').apply(lambda x: (x['Votes'].sum()))
weighted_rating = x/y
weighted_rating = weighted_rating.sort_values(ascending=False)

locality = np.array(weighted_rating.index[0:10])
rating = np.array(weighted_rating.values[0:10])

plt.figure(figsize=(15,7))
plt.bar(locality,rating,width=0.5)
plt.xticks(rotation=90)
plt.title('Locality vs Weighted rating',fontsize=20)
plt.xlabel('Locality',size=15)
plt.ylabel('Weighted rating',size=15)
plt.show()
```

# Problem 3.4 :
Plot the pie graph of top 10 cuisines present in restaurants in the USA.

```python
df = pd.read_csv('zomato.csv',encoding='ISO-8859-1')
df = df[df['Country Code']==216]
df.dropna(subset=['Cuisines'],inplace=True)

dic = {}
def func(x):
    menu = x.split(", ")
    for dish in menu:
        if dish in dic:
            dic[dish] = dic[dish] + 1
        else:
            dic[dish] = 1

df['Cuisines'].apply(lambda x: func(x))

keys = list(dic.keys())
values = list(dic.values())
value_index = np.argsort(values)
value_index = np.flip(value_index)[0:10]

cuisine = []
count = []
print('***** Top 10 Cuisines in USA *****')
for i in value_index:
    print(keys[i],values[i])
    cuisine.append(keys[i])
    count.append(values[i])

plt.pie(count,labels=cuisine,autopct='%.2f%%',radius=2)
plt.show()
```

# Problem 3.5 :
Plot the bubble graph of a number of Restaurants present in the city of India and keeping the weighted restaurant rating of the city in a bubble.

```python
df = pd.read_csv('zomato.csv',encoding='ISO-8859-1')
df = df[df['Country Code']==1]
df.dropna(subset=['City','Aggregate rating','Votes'],inplace=True)

x = df.groupby('City').apply(lambda x: (x['Votes']*x['Aggregate rating']).sum())
y = df.groupby('City').apply(lambda x: (x['Votes']).sum())
weighted_rating = x/y
city = df['City'].value_counts()

plt.figure(figsize=(15,7))
plt.scatter(city.index,city,s=weighted_rating*20)
plt.xticks(rotation=90)
plt.title('City with number of restaurants with weighted ratings',fontsize=20)
plt.xlabel('City',fontsize=15)
plt.ylabel('Number of Restaurants',fontsize=15)
plt.grid()
plt.show()
```
